```python
import os
import pathlib

dirpath = os.getcwd()

print(dirpath)


def listdirectory(dirpath:str,*,
                  extensions: list[str] = [],
                  excluded_folders: list[str] = []):
    
    # define folder exclusion strategy
    grab_all_folders = False
    if not excluded_folders:
        grab_all_folders = True
        
    # define extension strategy
    if not extensions:
        grab_all_extensions = True
    else:
        grab_all_extensions = False
        
        #checks extensions
        
        
    # initiate result
    result = set()

    # iterate over files
    for dir_, _, files in os.walk(dirpath):
        
        if not any(substring in dir_ for substring in excluded_folders) or grab_all_folders:       
            for file_name in files:
                if pathlib.Path(file_name).suffix in extensions or grab_all_extensions:
                    rel_dir = os.path.relpath(dir_, dirpath)
                    rel_file = os.path.join(rel_dir, file_name)
                    result.add(rel_file)

    return list(result)



a = listdirectory(dirpath,
                  extensions=["py"],
                  excluded_folders=["venv","work",".git","htmlcov","__pycache__"])


print(a)


```
